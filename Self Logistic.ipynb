```python
from sklearn import datasets
import sklearn
import numpy as np
import matplotlib.pyplot as plt

iris = sklearn.datasets.load_iris()
X = iris.data[:, :2]
y = (iris.target != 0) *1
X = np.c_[np.ones((X.shape[0], 1)), X]
y = y[:, np.newaxis]
theta = np.zeros((X.shape[1], 1))
lr = 0.5
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def cost(h, y):
    return (-y * np.log(h) - (1 - y) * np.log(1-h)).mean()

def gradient(X, y, h):
    return np.dot(X.T, h-y) / y.size

def predict_prob(X):
    return sigmoid(np.dot(X, theta))

def predict(X):
    return predict_prob(X) >= 0.5
```

```python



for i in range(1000):
    z = np.dot(X, theta)
    h = sigmoid(z)
    loss = cost(h,y)
    grad = gradient(X,y,h)
    theta -= (lr * grad)
    
    if i%10 == 0:
        data = (X.T[0], X.T[1])
        plt.scatter(X.T[1].T[:,np.newaxis], X.T[2].T[:,np.newaxis], c=y) 
        x_values = [4, 8]
        print(x_values)
        y_values = - (theta[0] + theta[1] * x_values) / theta[2]
        plt.plot(x_values, y_values)
        
        plt.show()
        print("Iteration: %d, Loss: %f" %(i,loss))
        



    
    
```
